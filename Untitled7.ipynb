```python
import pandas as pd
import re

def get_vcf(vcf_path):
    with open(vcf_path, "rt") as ifile:
          for line in ifile:
            if line.startswith("#CHROM"):
                  vcf_names = [x for x in line.split('\t')]
                  break
    ifile.close()
    return vcf_names


def vcf_edit(vcf_name):
  names = get_vcf(vcf_name)
  vcfDF = pd.read_csv(vcf_name, comment='#', delim_whitespace=True,header=None, names=names)
  vcfRest=vcfDF.loc[:,["#CHROM","POS","ID","REF","ALT"]]
  info = vcfDF['INFO\n'].str.split(';', expand=True)

  for row in info.itertuples():
    if 'ANN=' in row:
      ann = info[11]
    else: 
      print('ww')
      ann = info[12]
    

    copyOfann = ann.copy()
    for i in ann:
      if i.startswith('ANN'):
        ann_ext = ann.str.replace(',','\n',regex=True)
        ann_ext = ann_ext.str.split("|", expand=True)
        ann_ext = ann_ext.iloc[:,[0,3,4,6,9,10]]
        ann_ext.columns=["Nucleotide","Gene","GeneID","TranscriptID","Change","Aminoacid"]
        result = ann_ext["Aminoacid"].str.split('([A-Za-z]+)(\d+)([A-Za-z]+)', expand=True)
        result = result.loc[:,[1,3,2]]
        result.columns=["ReferenceAA","AlternativeAA","Codon"]
        Frames = [vcfRest, ann_ext, result]
        mergeDf = pd.concat([vcfRest, ann_ext, result], axis=1)
        mergeDf.head()
      
        return mergeDf
      
    
```

```python
import csv
 
lines =[]
with open("orj500_new4.vcf") as file:
    tsv_file = csv.reader(file, delimiter="\t")
    for line in tsv_file:
      if line[0][0] != '#':
        lines.append(line)
```

```python
updated_lines=[]
for line in lines: 
  indv_line=[]
  indv_line.append(line[0])
  indv_line.append(line[1])
  indv_line.append(line[2])
  indv_line.append(line[3])
  indv_line.append(line[4])
  seperated = line[7].split(';')
  for element in seperated: 
    if element.find('ANN') != -1: 
      annotation= element
  annotation_sep = annotation.split('|')
  indv_line.append(annotation_sep[1])
  indv_line.append(annotation_sep[2])
  indv_line.append(annotation_sep[3])
  indv_line.append(annotation_sep[4])
  indv_line.append(annotation_sep[6])
  indv_line.append(annotation_sep[7])
  indv_line.append(annotation_sep[9])
  indv_line.append(annotation_sep[10])
  updated_lines.append(indv_line)
```

```python
df = pd.DataFrame(updated_lines)
```
